```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
def tb(mystr):
    '''
    tb = triple brackets
    return the string with single brackets replaced 
    by triple brackets
    '''
    a = mystr.replace('{', '{{{')
    return a.replace('}','}}}')
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel('x axis')
ax1.set_ylabel('y axis')
ax1.set_title('Question about MathText')
x = np.linspace(1, 10, 10)
slope = .5
intercept = 1
y = 1/(slope*x+intercept)
ax1.plot(x,y)

sub1 = tb("$y = \\frac{}{}$")
sub2 = tb("$x = \\frac{}{}$")
sub3 = tb("\\frac{}{}")
ax1.text(4,.6, sub1.format(1,'{:.2f}*x + {:.2f}'.format(slope, intercept)), fontsize=22)
ax1.text(4,.4, sub2.format(sub3.format(1, 'y')+'-{:.2f}'.format(intercept), '{:.2f}'.format(slope)), fontsize=22)
```
